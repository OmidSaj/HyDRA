```python
def Advanced_MFCC_data(signal,sample_rate,
                       frame_size,frame_stride,
                       nfilt,NFFT,alpha,beta,filt_shape,
                       pre_emphasis,
                       MeanNormMFCC=False,MeanNormMSFB=True,hamming_filt=True):
    
    signal = numpy.append(signal[0], signal[1:] - pre_emphasis * signal[:-1])
    
    f = np.linspace(0,sample_rate/2,int(NFFT/2)+1)
    
    frame_length, frame_step = frame_size * sample_rate, frame_stride * sample_rate  # Convert from seconds to samples
    signal_length = len(signal)
    frame_length = int(round(frame_length))
    frame_step = int(round(frame_step))
    num_frames = int(numpy.ceil(float(numpy.abs(signal_length - frame_length)) / frame_step))  # Make sure that we have at least 1 frame

    pad_signal_length = num_frames * frame_step + frame_length
    z = numpy.zeros((pad_signal_length - signal_length))
    pad_signal = numpy.append(signal, z) # Pad Signal to make sure that all frames have equal number of samples without truncating any samples from the original signal

    indices = numpy.tile(numpy.arange(0, frame_length), (num_frames+1, 1)) + numpy.tile(numpy.arange(0, (num_frames+1) * frame_step, frame_step), (frame_length, 1)).T
    frames = pad_signal[indices.astype(numpy.int32, copy=False)]
    
    frames_nohm=np.array(frames)
    if hamming_filt:
        frames *= numpy.hamming(frame_length)


    mag_frames = numpy.absolute(numpy.fft.rfft(frames, NFFT))  # Magnitude of the FFT
    pow_frames = ((1.0 / NFFT) * ((mag_frames) ** 2))  # Power Spectrum

    h2595=2595
    low_freq_mel = 0
    high_freq_mel = (h2595 * np.log10(1 + (sample_rate / 2) / alpha))  # Convert Hz to Mel
    mel_points = numpy.linspace(low_freq_mel, high_freq_mel, nfilt + 2)  # Equally spaced in Mel scale
    hz_points = (alpha * (10**(mel_points / h2595) - 1))  # Convert Mel to Hz
    bin_f = numpy.floor((NFFT + 1) * hz_points / sample_rate)

    s_hv=-4*beta/sample_rate # Decay/Increase slop
    x_mid=f[-1]/2
    y_mid=1
    mod_fact=np.ones((nfilt))
    print(mod_fact.shape)
    print(NFFT)
    print(f[bin_f.astype(int)].shape)
    mod_fact=(f[bin_f.astype(int)]-x_mid)*s_hv+1

    fbank = numpy.zeros((nfilt, int(numpy.floor(NFFT / 2 + 1))))
    for m in range(1, nfilt + 1):
        f_m_minus = int(bin_f[m - 1])   # left
        f_m = int(bin_f[m])             # center
        f_m_plus = int(bin_f[m + 1])    # right

        for k in range(f_m_minus, f_m):
            if filt_shape=='Tri':
                fbank[m - 1, k] = (k - bin_f[m - 1]) / (bin_f[m] - bin_f[m - 1])
            if filt_shape=='Rec':
                fbank[m - 1, k] = 1

        for k in range(f_m, f_m_plus):
            if filt_shape=='Tri':
                fbank[m - 1, k] = (bin_f[m + 1] - k) / (bin_f[m + 1] - bin_f[m])
            if filt_shape=='Rec':
                fbank[m - 1, k] = 1

    for i in range(fbank.shape[0]):
        fbank[i,:]=fbank[i,:]*mod_fact[i+1]
        
    filter_banks = numpy.dot(pow_frames, fbank.T)
    filter_banks = numpy.where(filter_banks == 0, numpy.finfo(float).eps, filter_banks)  # Numerical Stability
    filter_banks = numpy.log10(filter_banks)  # dB
    
    # checking with librosa
#     mfcc=librosa.feature.mfcc(y=signal, sr=sample_rate, dct_type=2, n_mfcc=nfilt,  hop_length=100, n_fft=NFFT)
#     mfcc=np.swapaxes(mfcc,0,1)

    mfcc = dct(filter_banks, type=2, axis=1, norm='ortho')
    print(mfcc.shape)
    if MeanNormMFCC:
        filter_banks-= (numpy.mean(filter_banks, axis=0) + 1e-8)
    if MeanNormMSFB:        
        mfcc-= (numpy.mean(mfcc, axis=0) + 1e-8)
    
    return mfcc, pow_frames,f,fbank,frames,filter_banks,bin_f,mod_fact,pad_signal,frames_nohm
```
